```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

file = ''
cp_raw = pd.read_excel(file, sheet_name=3)
cp_raw.head()
```

```python
mask = ((cp_raw['Time']>=2099) & (cp_raw['Time']<=2281) & (cp_raw['Pace']<2.5))
cp_raw = cp_raw[mask]
display(cp_raw)
```

```python
avg_power=cp_raw['Load'].mean()
cp=cp_raw[cp_raw['Time']>=2250]['Load'].mean()
w_prime=(avg_power-cp)*180
cp_data = pd.DataFrame({'Avg P(W)':[avg_power], 'CP(W)':[cp], 'Wprime(J)':[w_prime]})
display(cp_data)
```

```python
cp_data.to_excel('ES04_3minOA_PLA.xlsx')
```
