```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

file = '/content/ES_05_5step_3min_Lode_PLA.xlsx'
cp_raw = pd.read_excel(file, sheet_name=3)
display(cp_raw)
```

```python
linear_factor = cp_raw[(cp_raw['Time']>=2101) & (cp_raw['Time']<=2102)]['Linear Alpha']
display(linear_factor)
```

```python
linear_alpha = 0.029
tolerance = 1e-5
mask = ((cp_raw['Time']>=2099) & (cp_raw['Time']<=2281) & (abs(cp_raw['Linear Alpha'] - linear_alpha) < tolerance))
cp_raw = cp_raw[mask]
display(cp_raw)
```

```python
avg_power=cp_raw['Load'].mean()
cp=cp_raw[cp_raw['Time']>=2250]['Load'].mean()
w_prime=(avg_power-cp)*180
cp_data = pd.DataFrame({'Avg P(W)':[avg_power], 'CP(W)':[cp], 'Wprime(J)':[w_prime]})
display(cp_data)
```

```python
cp_data.to_excel('ES05_3minOA_PLA.xlsx')
```
